# Questions 2:
How to create a series from a numpy array?

```python
# importing Pandas & numpy 
import pandas as pd 
import numpy as np 

# numpy array 
data = np.array(['a', 'b', 'c', 'd', 'e']) 

# creating series 
s = pd.Series(data) 
print(s) 
```
